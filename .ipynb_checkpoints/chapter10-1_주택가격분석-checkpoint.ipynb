```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
boston = load_boston() #데이터셋 로듷여 boston 객체 생성
```

```python
print(boston.DESCR) #데이터셋에 대한 설명 확인
```

```python
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df.head() #출력하여 확인
```

```python
boston_df['PRICE']=boston.target #PRICE 칼럼을 추가
boston_df.head() #출력하여 확인
```

```python
print('보스톤 주택 가격 데이터셋 크기 : ',boston_df.shape) #행의 개수와 열의 개수 확인
```

```python
#행 506개 : 데이터 506개
#열 14개 : 변수 14개 - 13개의 독립변수X와 1개의 종속변수Y(PRICE)
```

```python
boston_df.info() #boston_df에 대한 정보 확인
```

#선형회귀를 이용해 분석 모델 구축하기

```python
from sklearn.linear_model import LinearRegression #회기분석
from sklearn.model_selection import train_test_split #데이터셋 분리 작업
from sklearn.metrics import mean_squared_error, r2_score #성능측정을 위한 평가지표
```

```python
# X, Y 분할하기
Y = boston_df['PRICE']
X = boston_df.drop(['PRICE'], axis=1, inplace=False) #PRICE를 제외한(drop) 나머지 피처
```

```python
#훈련용(학습용) 데이터와 평가용 데이터 분할하기 (7:3비율로 분할)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.3,
                                                   random_state = 156)
```

```python
#선형회귀분석 : 모델 생성
lr = LinearRegression()
```

```python
#선형회귀분석 : 모델훈련
lr.fit(X_train, Y_train)
```

```python
#선형회귀분석 : 평가데이터에 대한 예측 수행 -> 예측결과 Y_predict 구하기
Y_predict = lr.predict(X_test)
```

```python
#결과값(Y_test)과 예측결과(Y_predict)의 오차를 계산
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict)))
```

```python
print('Y 절편값: ',  lr.intercept_)
print('회귀 계수 값: ', np.round(lr.coef_, 1))
```

```python
#회귀 계수 값, 피처이름을 묶어 Series 자료형으로 만들고 -> 회귀 계수 값 기준으로 내림차순 정렬해 확인
coef = pd.Series(data = np.round(lr.coef_,2),index=X.columns)
coef.sort_values(ascending=False)
```

#회귀 분석 결과를 산점도 + 선형 회귀 그래프로 시각화하기

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axs = plt.subplots(figsize = (16,16), ncols=3, nrows=5)

x_features = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE',
             'DIS','RAD','TAX','PTRATIO','B','LSTAT']

for i,feature in enumerate(x_features):
    row = int(i/3)
    col = i%3
    sns.regplot(x=feature, y='PRICE', data=boston_df,
               ax=axs[row][col])
```
