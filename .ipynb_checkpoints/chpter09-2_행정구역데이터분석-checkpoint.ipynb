# 행정구역별 공공보건의료기관의 수 구하기

```python
#시도군구 컬럼 추가
addr['시도군구'] = addr.apply(lambda r: r['시도'] + '' + r['군구'],axis = 1)
addr.head()
```

```python
addr['count']=0
addr_group = pd.DataFrame(addr.groupby(['시도','군구','시도군구'],as_index=False)
                          .count()) #그룹을 만들어, 그룹별 원소 개수를 count에 저장
addr_group.head() #출력하여 내용 확인
```

```python
#데이터 병합에 사용할 인덱스 설정
addr_group = addr_group.set_index("시도군구")
addr_group.head() #작업내용확인
```

# 행정구역별 인구수 데이터 준비하기

```python
#데이터 불러오기
population = pd.read_excel('9-2_data\\행정구역_시군구_별__성별_인구수_2.xlsx')
population.head() #작업 확인용 출력
```

```python
#컬럼이름변경
population = population.rename(columns = {'행정구역(시군구)별(1)':'시도',
                                         '행정구역(시군구)별(2)':'군구'})
population.head() #작업확인용출력
```

#시도군구 컬럼 만들기

```python
#문자열 앞뒤에 포함된 띄어쓰기 공백 모두 제거 strip()
for element in range(0,len(population)):
    population['군구'][element]=population['군구'][element].strip()
    
#시도군구 컬럼에 값 연결한거 추가
population['시도군구']=population.apply(lambda r: r['시도']+''+
                                   r['군구'],axis=1)

population.head() #작업확인용출력
```

```python
# '소계'행은 필요 없으므로 제외
population = population[population.군구 != '소계']
population.head() #작업확인용출력
```

```python
#시도군구 컬럼을 데이터 병합에 사용할 인덱스로 설정
population = population.set_index("시도군구")
population.head() #작업 확인용 출력
```

```python
#addr_group과 population을 내부병합응로 병합
addr_population_merge = pd.merge(addr_group,population, how='inner',
                                left_index=True, right_index=True)
addr_population_merge.head() #작업 확인용 출력
```

```python
#필요한 컬럼만 추출해 local_MC_Population 만들기
local_MC_Population = addr_population_merge[['시도_x','군구_x',
                                             'count','총인구수 (명)']]
local_MC_Population.head() #작업확인용출력
```

```python
#컬럼 이름 변경
local_MC_Population = local_MC_Population.rename(
    columns = {'시도_x':'시도','군구_x':'군구','총인구수 (명)':'인구수'})

#인구수 대비 공공보건의료기관 비율 구해 컬럼 추가
MC_count = local_MC_Population['count']
local_MC_Population['MC_ratio'] = MC_count.div(local_MC_Population['인구수'],
                                             axis = 0)*100000

local_MC_Population.head() #작업확인용출력
```

#블록맵으로 시각화하기

```python
#행정구역의 블록 위치 x,y 데이터를 정의
BORDER_LINES = [
    [(3, 2), (5, 2), (5, 3), (9, 3), (9, 1)], # 인천
    [(2, 5), (3, 5), (3, 4), (8, 4), (8, 7), (7, 7), (7, 9), (4, 9), (4, 7), (1, 7)], # 서울
    [(1, 6), (1, 9), (3, 9), (3, 10), (8, 10), (8, 9),
     (9, 9), (9, 8), (10, 8), (10, 5), (9, 5), (9, 3)], # 경기도
    [(9, 12), (9, 10), (8, 10)], # 강원도
    [(10, 5), (11, 5), (11, 4), (12, 4), (12, 5), (13, 5),
     (13, 4), (14, 4), (14, 2)], # 충청남도
    [(11, 5), (12, 5), (12, 6), (15, 6), (15, 7), (13, 7),
     (13, 8), (11, 8), (11, 9), (10, 9), (10, 8)], # 충청북도
    [(14, 4), (15, 4), (15, 6)], # 대전시
    [(14, 7), (14, 9), (13, 9), (13, 11), (13, 13)], # 경상북도
    [(14, 8), (16, 8), (16, 10), (15, 10),
     (15, 11), (14, 11), (14, 12), (13, 12)], # 대구시
    [(15, 11), (16, 11), (16, 13)], # 울산시
    [(17, 1), (17, 3), (18, 3), (18, 6), (15, 6)], # 전라북도
    [(19, 2), (19, 4), (21, 4), (21, 3), (22, 3), (22, 2), (19, 2)], # 광주시
    [(18, 5), (20, 5), (20, 6)], # 전라남도
    [(16, 9), (18, 9), (18, 8), (19, 8), (19, 9), (20, 9), (20, 10)], # 부산시
]
```

```python
import numpy as np
```

```python
#블록맵의 블록에 데이터 매핑, 색표시해 블록맵 그려 저장하는 함수 정의

def draw_blockMap(blockedMap, targetData, title, color ):
    whitelabelmin = (max(blockedMap[targetData]) - min(blockedMap[targetData])) * 0.25 + min(blockedMap[targetData])

    datalabel = targetData

    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])

    mapdata = blockedMap.pivot(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(8, 13))
    plt.title(title)
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=color, edgecolor='#aaaaaa', linewidth=0.5)

    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():
        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'
    
        # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다. (중구, 서구)
        if row['광역시도'].endswith('시') and not row['광역시도'].startswith('세종'):
            dispname = '{}\n{}'.format(row['광역시도'][:2], row['행정구역'][:-1])
            if len(row['행정구역']) <= 2:
                dispname += row['행정구역'][-1]
        else:
            dispname = row['행정구역'][:-1]

        # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시한다.
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 9.5, 1.5
        else:
            fontsize, linespacing = 11, 1.2

        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=fontsize, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)
    
    # 시도 경계 그린다.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=4)

    plt.gca().invert_yaxis()
    #plt.gca().set_aspect(1)
    plt.axis('off')
    
    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)

    plt.tight_layout()
    
    plt.savefig('.\\9장_data\\' + 'blockMap_' + targetData + '.png')
                      
    
    plt.show()
```

```python
#count 컬럼 값에 대해 Blues 색상 스펙트럼 사용해 블록맵 작성
draw_blockMap(data_draw_korea_MC_Population_all, 'count', '행정구역별 공공보건의료기관 수', 'Blues')
```

```python
draw_blockMap(data_draw_korea_MC_Population_all, 'MC_ratio', '행정구역별 인구수 대비 공공보건의료기관 비율', 'Reds' )
```
